```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def avg_data(year):
    year = int(year)
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi20{}.csv'.format(year), chunksize=24):
        add_var = 0
        avg = 0.0
        data = []
        df = pd.DataFrame(data=rows)
        for index,row in df.iterrows():
            data.append(row['PM2.5'])
        for i in data:
            if type(i) is float or type(i) is int:
                add_var = add_var + i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and  i!='---' and i!='InVld':
                    temp = float(i)
                    add_var = add_var + temp
        avg = add_var/24
        temp_i = temp_i + 1
        average.append(avg)
    return average
```

```python
#plt.rcdefaults()
plt.style.use('seaborn')
list1 = [13,14,15,16,17,18]
for i in list1:
    lst = avg_data(i)
    if i!=14 and i!=18:
        plt.plot(range(0,365), lst, label="20{} data".format(i) ,color='coral', linestyle='-')
    else:
        plt.plot(range(0,364), lst, label="20{} data".format(i) ,color='coral', linestyle='-')#sns.lineplot(x=range(0,364), y=lst, label="20{} data".format(i), color="coral")
    plt.title("Air Pollution 20{}".format(i) ,size=15)
    plt.xlabel('Day',fontweight='bold')
    plt.ylabel('PM 2.5',fontweight='bold')
    plt.legend(loc='upper right')
    plt.show()
```
